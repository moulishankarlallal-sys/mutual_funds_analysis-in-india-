```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
funds = {
    'Fund Name':[
        'SBI Balanced Advantage',
        'ICICI Prudential Balanced Advantage',
        'HDFC Balanced Advantage',
        'Kotak Balanced Advantage',
        'Nippon India Balanced Advantage'
    ],

    '1Y Return':[12.5,14.2,13.8,11.9,13.1],

    '3Y Return':[16.8,18.5,17.6,15.9,17.1],

    '5Y Return':[15.9,17.2,16.8,15.1,16.3],

    'Expense Ratio':[0.82,0.79,0.91,0.85,0.89],

    'AUM(Cr)':[32000,58000,69000,18000,25000],

    'Standard Deviation':[12.5,11.2,12.1,11.9,12.3],

    'Beta':[0.89,0.82,0.91,0.87,0.88],

    'Alpha':[2.4,3.1,2.8,2.1,2.6]
}

df = pd.DataFrame(funds)

df

risk_free_rate = 6

df['Sharpe Ratio'] = (df['5Y Return']-risk_free_rate)/df['Standard Deviation']

df

df['Rank(Return)']=df['5Y Return'].rank(ascending=False)

df['Rank(Sharpe)']=df['Sharpe Ratio'].rank(ascending=False)

df['Overall Rank']=(df['Rank(Return)']+df['Rank(Sharpe)'])/2

df=df.sort_values('Overall Rank')

df

best=df.iloc[0]

print("Top Performing Mutual Fund")

print(best[['Fund Name','5Y Return','Sharpe Ratio','Overall Rank']])

plt.figure(figsize=(10,5))

plt.bar(df['Fund Name'],df['5Y Return'])

plt.title("5-Year Return Comparison")

plt.ylabel("Return (%)")

plt.xticks(rotation=20)

plt.show()

plt.figure(figsize=(10,5))

plt.bar(df['Fund Name'],df['Expense Ratio'])

plt.title("Expense Ratio Comparison")

plt.ylabel("Expense Ratio")

plt.xticks(rotation=20)

plt.show()

plt.figure(figsize=(10,5))

plt.bar(df['Fund Name'],df['AUM(Cr)'])

plt.title("Assets Under Management")

plt.ylabel("AUM (Crores)")

plt.xticks(rotation=20)

plt.show()

plt.figure(figsize=(8,6))

plt.scatter(df['Standard Deviation'],
            df['5Y Return'],
            s=250)

for i in range(len(df)):
    plt.text(df['Standard Deviation'][i],
             df['5Y Return'][i],
             df['Fund Name'][i])

plt.xlabel("Risk (Standard Deviation)")

plt.ylabel("Return (%)")

plt.title("Risk vs Return")

plt.show()

plt.figure(figsize=(10,5))

plt.bar(df['Fund Name'],df['Sharpe Ratio'])

plt.title("Sharpe Ratio Comparison")

plt.ylabel("Sharpe Ratio")

plt.xticks(rotation=20)

plt.show()

plt.figure(figsize=(10,5))

plt.bar(df['Fund Name'],df['Alpha'])

plt.title("Alpha Comparison")

plt.ylabel("Alpha")

plt.xticks(rotation=20)

plt.show()

plt.figure(figsize=(10,5))

plt.bar(df['Fund Name'],df['Beta'])

plt.title("Beta Comparison")

plt.ylabel("Beta")

plt.xticks(rotation=20)

plt.show()

df['Score']=(
    df['5Y Return']*0.35+
    df['Sharpe Ratio']*15+
    df['Alpha']*5-
    df['Expense Ratio']*10-
    df['Beta']*2
)

df=df.sort_values("Score",ascending=False)

print(df[['Fund Name','Score']])

df.to_excel("Competitive_Analysis_Mutual_Funds.xlsx",
            index=False)

print("Excel File Saved Successfully")

```
